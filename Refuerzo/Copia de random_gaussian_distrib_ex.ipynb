```python
import numpy as np

# Datos del código anterior
o = np.array([[3., 2., 2.],
              [2., 3., 2.]])

# Inicializar la media y el contador
media = 0
n = 0

# Listas para almacenar las medias y los pasos
medias = []
pasos = []

# Calcular la media con la fórmula de la media actualizada
for valor in o[:,0]:
  n += 1
  media = media + (valor - media) / n
  medias.append(media)
  pasos.append(n)

# Graficar la evolución de la media
import matplotlib.pyplot as plt
plt.plot(pasos, medias, marker='o')
plt.xlabel('Pasos')
plt.ylabel('Media')
plt.title('Evolución de la media')
plt.show()
```
